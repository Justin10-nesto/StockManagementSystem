```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
regions = {'region':[],'link':[]}
```

```python
URL = "https://matokeo.necta.go.tz/psle_selection.htm"
```

```python
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
```

```python
table = soup.find('table') 
```

```python
for row in table.findAll('td'):
    full_link_tag = row.find('p').find('a')
    try:
        link = full_link_tag['href']
        regions['region'].append(full_link_tag.text)
        regions['link'].append(URL[:-18] +link)
        print(link)
    except:
        pass
```

```python
districts ={ 'district':[], 'link':[]}
```

```python
for region in regions['link'][:5]:
    r = requests.get(region)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
    table = soup.find('table') 
    for row in table.findAll('td'):
        full_link_tag = row.find('p').find('a')
        try:
            link = full_link_tag['href']
            districts['district'].append(full_link_tag.text)
            districts['link'].append(region[:-10] +link)
            print(link)
        except:
            pass
```

```python
districts
```

```python
schools = {'school':[], 'link':[]}
```

```python
for district in districts['link']:
    r = requests.get(district)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
    table = soup.find('table') 
    for row in table.findAll('td'):
        full_link_tag = row.find('p').find('a')
        try:
            link = full_link_tag['href']
            schools['school'].append(full_link_tag.text)
            schools['link'].append(district[:-14] +link)
            print(link)
        except:
            pass
```

```python
selected_schools = []
```

```python
for school in schools['link']:
    ttable_MN = pd.read_html(school)
    selected_schools.append(ttable_MN)
```

```python
table_MN = pd.read_html(schools['link'][0])
```

```python
default_school =table_MN[0]
```

```python
for sl_school in selected_schools:
    default_school = pd.concat([default_school, sl_school[0]])
```

```python
default_school.drop_duplicates(inplace= True)
```

```python
default_school.to_csv('selected_standard7.csv')
```
